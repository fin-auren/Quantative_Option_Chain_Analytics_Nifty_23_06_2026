```python
def clean_numeric(series):
    return (
        series.astype(str)
        .str.replace(",", "", regex=False)
        .replace("nan", np.nan)
        .replace("None", np.nan)
    )
```

```python
num_cols = [
    "call_oi","call_change_oi","call_volume",
    "call_iv","call_ltp","call_change",
    "put_bid_qty","put_bid","put_ask","put_ask_qty",
    "strike",
    "put_change","put_ltp","put_iv","put_volume",
    "put_change_oi","put_oi"
]

for col in num_cols:
    df[col] = clean_numeric(df[col])
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
import pandas as pd

metrics = [
    "call_oi",
    "put_oi",
    "call_change_oi",
    "put_change_oi",
    "call_volume",
    "put_volume",
    "call_iv",
    "put_iv"
]

interpretation_map = {
    "call_oi": "Potential Resistance",
    "put_oi": "Potential Support",
    "call_change_oi": "Maximum New Call Positions",
    "put_change_oi": "Maximum New Put Positions",
    "call_volume": "Most Active Call Trading",
    "put_volume": "Most Active Put Trading",
    "call_iv": "Highest Call Implied Volatility",
    "put_iv": "Highest Put Implied Volatility"
}

summary_rows = []

for metric in metrics:

    idx = df[metric].idxmax()

    strike = df.loc[idx, "strike"]

    value = df.loc[idx, metric]

    summary_rows.append({
        "Metric": metric,
        "Strike": strike,
        "Value": value,
        "Interpretation": interpretation_map[metric]
    })

summary_df = pd.DataFrame(summary_rows)

print(summary_df)
```

```python
Total_put_OI = df["put_oi"].sum()
Total_call_OI = df["call_oi"].sum()
Total_put_volume = df["put_volume"].sum()
Total_call_volume = df["call_volume"].sum()

PCR = Total_put_OI / Total_call_OI
PCR

def pcr_anal(pcr):
  if pcr < 0.7:
    return print("Market is bearish. PCR : ",pcr)
  elif 0.7 < pcr < 1.0:
    return print("Market is slighlty bearish. PCR : ",pcr)
  elif 1.0 < pcr < 1.3:
    return print("Market is neutral. PCR : ",pcr)
  elif 1.3 < pcr < 1.5:
    return print("Market is Bullish. PCR : ",pcr)
  else:
    return print("Market is Extremely Bullish. PCR : ",pcr)

pcr_anal(PCR)

```

```python
weighted_mean_strike = (
    (df["strike"] * df["call_oi"]).sum()
    / df["call_oi"].sum()
)
print(weighted_mean_strike)
```

```python
temp = df[["strike", "call_oi"]].dropna()
weighted_variance = ((temp["call_oi"] * (temp["strike"] - weighted_mean_strike)**2).sum()
) / temp["call_oi"].sum()

print("Weighted Variance :", weighted_variance)

import numpy as np

weighted_std = np.sqrt(weighted_variance)

print("Weighted Standard Deviation : ", weighted_std)

df2 = pd.DataFrame({
    "Metric":["Wighted Variance","Weighted Standard Deviation"],
    "Value":[weighted_variance,weighted_std]
})

df2
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------
# Create clean IV dataframe
# ---------------------------------------

iv_df = df[
    ["strike", "call_iv", "put_iv"]
].copy()

iv_df = iv_df.dropna()

iv_df = iv_df.sort_values("strike")

# ---------------------------------------
# IV Smile Plot
# ---------------------------------------

plt.figure(figsize=(12,6))

plt.plot(
    iv_df["strike"],
    iv_df["call_iv"],
    marker='o',
    markersize=3,
    linewidth=2,
    label="Call IV"
)

plt.plot(
    iv_df["strike"],
    iv_df["put_iv"],
    marker='s',
    markersize=3,
    linewidth=2,
    label="Put IV"
)

plt.xlabel("Strike Price")
plt.ylabel("Implied Volatility (%)")
plt.title("NIFTY Option Chain Implied Volatility Smile")
plt.grid(True, alpha=0.3)
plt.legend()

plt.show()

# ---------------------------------------
# Summary Statistics
# ---------------------------------------

iv_summary = pd.DataFrame({
    "Metric": [
        "Mean IV",
        "Median IV",
        "Std IV",
        "Min IV",
        "Max IV"
    ],

    "Call IV": [
        iv_df["call_iv"].mean(),
        iv_df["call_iv"].median(),
        iv_df["call_iv"].std(),
        iv_df["call_iv"].min(),
        iv_df["call_iv"].max()
    ],

    "Put IV": [
        iv_df["put_iv"].mean(),
        iv_df["put_iv"].median(),
        iv_df["put_iv"].std(),
        iv_df["put_iv"].min(),
        iv_df["put_iv"].max()
    ]
})

print("\nIV Summary Statistics")
print(iv_summary)

# ---------------------------------------
# Maximum IV Locations
# ---------------------------------------

call_iv_idx = iv_df["call_iv"].idxmax()
put_iv_idx  = iv_df["put_iv"].idxmax()

max_iv_summary = pd.DataFrame({

    "Metric": [
        "Maximum Call IV",
        "Maximum Put IV"
    ],

    "Strike": [
        iv_df.loc[call_iv_idx, "strike"],
        iv_df.loc[put_iv_idx, "strike"]
    ],

    "IV": [
        iv_df.loc[call_iv_idx, "call_iv"],
        iv_df.loc[put_iv_idx, "put_iv"]
    ]
})

print("\nMaximum IV Summary")
print(max_iv_summary)

# ---------------------------------------
# Top 5 Highest IV Strikes
# ---------------------------------------

print("\nTop 5 Call IV Strikes")
print(
    iv_df.nlargest(
        5,
        "call_iv"
    )[["strike","call_iv"]]
)

print("\nTop 5 Put IV Strikes")
print(
    iv_df.nlargest(
        5,
        "put_iv"
    )[["strike","put_iv"]]
)
```
